# Subplots

```python
import matplotlib.pyplot as plt
import seaborn as sns

from math import ceil
```

```python
data = sns.load_dataset("iris")
data
```

```python
ncol = 10
nrow = 5
print(f"{ncol=}, {nrow=}, {nrow*ncol=}")
```

```python
## suplots
fig, axs = plt.subplots(nrow, ncol ,figsize=(15,5))
axs = axs.reshape(-1)       ## flat axes
for i,ax in enumerate(axs):
    plt.text(0.5, 0.5
            , str(i)
            , horizontalalignment='center',verticalalignment='center'
            , transform=ax.transAxes
        )
    ax.set_xticks([])
    ax.set_yticks([])
plt.tight_layout()
plt.show()
```

Remove subplots

```python
## suplots
fig, axs = plt.subplots(nrow, ncol ,figsize=(15,5))
axs = axs.reshape(-1)       ## flat axes
for i,ax in enumerate(axs):
    plt.text(0.5, 0.5
            , str(i)
            , horizontalalignment='center',verticalalignment='center'
            , transform=ax.transAxes
        )
    ax.set_xticks([])
    ax.set_yticks([])
# subplots to remove
for i in range(0,ncol*nrow,3):
    axs[i].axis('off')
fig.suptitle("remove every 3 subplots", fontsize=10)
plt.tight_layout()
plt.show()
```

set label to only one subplot

```python
## suplots
fig, axs = plt.subplots(nrow, ncol ,figsize=(15,5))
axs = axs.reshape(-1)       ## flat axes
for i,ax in enumerate(axs):
    plt.text(0.5, 0.5
            , str(i)
            , horizontalalignment='center',verticalalignment='center'
            , transform=ax.transAxes
        )
    sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species", ax=ax)
    ax.set_xticks([])
    ax.set_yticks([])
fig.suptitle("remove every 3 subplots", fontsize=10)
plt.tight_layout()
plt.show()
```
